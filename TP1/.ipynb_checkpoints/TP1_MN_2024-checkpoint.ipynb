### Boucles et logique en Python

La boucle **FOR** en Python s'écrit de la façon suivante:

```python
for k in range(5):
    print(k) # visualization de la variable
```

Il y a plusieurs commentaires à faire ici:

- la boucle affiche les numéros de 0 à 4, alors que l'on aurait pu s'attendre aux numéros de 1 à 5! Ceci est dû au fait (très important!) qu'en Python, les indices des listes, tableaux et autres "itérables" commencent à 0.
- Grosso modo, `range(5)` renvoie une liste `[0,1,2,3,4]`. Donc la première ligne veut simplement dire "pour k dans [0,1,2,3,4]". 
- la première ligne se termine par `:` , il faut le mettre après chaque expression de ce type (voir plus bas)
- la deuxième ligne (intérieur de la boucle) doit être décalée! Cela peut être un décalage d'une tabulation, 2 espaces, 4 espaces, etc. Choisissez ce que vous voulez, l'important c'est d'être consistant. La communauté recommande d'utiliser une tabulation = 4 espaces.

La condition **IF** en Python s'écrit de la façon suivante :

```python
if 2 > 0 and 6 > 4:
    print("Ces deux propriétés sont vraies")
```

```python
if 2 != 0 and 6 == 4:
    print("Ces deux propriétés sont vraies")
else:
    print("Une de ces propriétés au moins est fausse")
```

- Tout comme la boucle for, chaque "phrase" doit se terminer par un `:`. Cela concerne la ligne du `if`, mais aussi celle du `else`, si il y en a.
- On indente toujours les instructions à l'intérieur du `if` ou `else`
- On peut combiner plusieurs propositions syntaxiques avec `and` et `or`. On peut aussi utiliser des parenthèses pour les regrouper si besoin

On peut évidemment combiner tout ceci:

```python
for k in range(5):
    if k in [0,2,4]:
        print("k est pair")
    else:
        print("k est impair")
```

### Exercice 3 : Modèle de Lotka-Volterra

On considère le système suivant, utilisé comme modèle chimique (réactions), écologique (proie-prédateur) et économique :

$$

	        x'(t) = x(t) \left( 1 - y(t) \right) \\
	        y'(t) = y(t) \left( x(t) - 1 \right)

	$$
    
avec  $x(0) = x_0$ et $y(0) = y_0$.

**(1)** Vérifiez que pour tout $(x_0, y_0)$, le système admet une solution locale.

**(2)** Vérifiez que si $t \mapsto (x(t), y(t))$ est une solution, alors la fonction

**(3)** Ecrivez une fonction `f(x,y)` qui au point $(x,y)$ fait correspondre le vecteur $(x',y')$.

```python
def f(x,y):
    return (x * (1-y),y * (x - 1))
```

**(4)** Tracez un graphe de l'espace des phases du système : dans le plane $(x,y)$, construire et tracer le champ des vecteurs

### Exercice 4 : Méthodes d'approximations successives

On s'intéresse à résoudre l'équation différentielle suivante 
$$ 
y' = 2t(1+y)
$$
en utilisant sa forme intégale,
$$
\phi(t) = \int_0^t 2s(1+\phi(s)) \,ds.
$$

**(1)** Écrivez une fonction `RecurrencePhi(phi0 ,T, N)` qui prend en entrée une fonction sous la forme d'un vecteur `phi0`, un temps final `T` et `N` le nombre de points dans la discrétisation de l'intervalle $[0,T]$ et qui renvoie la fonction, $\phi_1$, définie sur $[0,T]$ donnée par une étape de de la suite
$$
\phi_1(t) = \int_0^t 2s(1+\phi_0(s)) \, ds.
$$

Pour calculer l'intégrale, on pourra l'approcher par une somme de Riemann,
$$
 \int_0^t 2s(1+\phi(s)) \,ds \approx h\sum_{k = 0}^{n} 2kh(1+\phi(kh))
$$
où $nh = t$ et $n\leq N$.

```python
def RecurrencePhi(phi0 , t):
    
    phi = phi0
    h = t[1]-t[0]
    return np.cumsum(h * 2 * t * (1 + phi0))
```

```python
RecurrencePhi(np.zeros(100),np.linspace(0,2,100))
```

**(2)** Utilisez la fonction `RecurrencePhi(phi0 , T, N)` pour calculer une solution approchée de l'équation différentielle avec `phi0`la fonction nulle, `N=1000`, `T = 2`, après `nmax` itérations. Tracer le graphe de $\phi_{nmax}$. 

```python
def phi0(t,nmax):
    phi = np.zeros_like(t)
    for i in range(nmax):
        phi = RecurrencePhi(phi,t)
    return phi
```

```python
t = np.linspace(0,2,1000)
plt.plot(t, phi0(t,4))
plt.plot(t, np.exp(t**2) - 1,"r:")
```

**(3)** Comparez la solution obtenue à la solution exacte $y(t) = e^{t^2} -1$. 

```python
t = np.linspace(0,2,1000)
for i in [2,4,8,10,50]:
    plt.plot(t, np.abs(phi0(t,i)- np.exp(t**2) - 1),label=f"{i} itérations")

plt.legend()
plt.title("erreur absolue")
```
